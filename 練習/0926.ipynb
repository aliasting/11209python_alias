```python
import random
import pyinputplus as pyip

def diceGame():    
    j = 0
    t = 0
    score_t = 0 

    num = [random.randint(1,6) for i in range(4) ]  
    #num = [2,2,6,6]
    print(num,end=" ")
    for x in range(4):
        for j in range(x+1,4) :
            if num[x] == num[j] :
                t += 1
                score_t += num[j]            
    if t == 1 :
        point = sum(num)-score_t*2
        return point
    elif  t == 2:
        point = int(max(num)*2)
        return point
    elif  t == 6:
        point = int(sum(num)*10)
        return point
    else :
        point = 0
        return point 


while True:
    print()
    print("===============擲骰子遊戲=================:\n")     
    a = diceGame()
    print(f"電腦擲骰子點數:{a}點")    
    b = diceGame()    
    print(f"您擲骰子點數:{b}點")
    if b > a:
        print("you win") 
    else :
        print("you lose")
    answer = pyip.inputYesNo(prompt="還要繼續嗎?(y,n)",yesVal='y',noVal='n')
    if not (answer == 'y'):
        print()
        break
print("遊戲結束")
```

```python
import random
class Player:
    def __init__(self,n:str):
        self.name = n
        self.__dice1 = random.randint(1,6)
        self.__dice2 = random.randint(1,6)
        self.__dice3 = random.randint(1,6)
        self.__dice4 = random.randint(1,6)

    #method
    def __play(self) -> int:
        while True:
            dice = [self.__dice1,self.__dice2,self.__dice3,self.__dice4]
            if dice[0] == dice[1] == dice[2] == dice[3]:   
                score = 12 + self.__dice1
            elif dice[0] == dice[1] == dice[2] or dice[0] == dice[1] == dice[3]:   
                score = 0
            elif dice[0] == dice[2] == dice[3] or dice[1] == dice[2] == dice[3]:   
                score = 0
            
            elif dice[0] == dice[1] and dice[2] == dice[3]:    
                if dice[0] > dice[2]:
                    score = dice[0] + dice[1]
                else:
                    score = dice[2] + dice[3]
            elif dice[0] == dice[2] and dice[1] == dice[3]:    
                if dice[0] > dice[1]:
                    score = dice[0] + dice[2]
                else:
                    score = dice[1] + dice[3]

            elif dice[0] == dice[1]:    
                score = dice[2] + dice[3]
            elif dice[0] == dice[2]:    
                score = dice[1] + dice[3]
            elif dice[0] == dice[3]:    
                score = dice[1] + dice[2]
            elif dice[1] == dice[2]:    
                score = dice[0] + dice[3]
            elif dice[1] == dice[3]:    
                score = dice[0] + dice[2]
            elif dice[2] == dice[3]:    
                score = dice[0] + dice[1]
            elif not (dice[0] == dice[1] == dice[2] == dice[3]):   
                score = 0
            return score

    @property
    def value(self) -> int:
        return self.__play()
        
    def __repr__(self) -> str:
        descript = ""
        descript += f"{self.name}\n"
        descript += f"骰子1={self.__dice1}\n"
        descript += f"骰子2={self.__dice2}\n"
        descript += f"骰子3={self.__dice3}\n"
        descript += f"骰子4={self.__dice4}\n"
        descript += f"得分={self.value}分\n"
        return descript
if __name__ == '__main__':
    p1 = Player("rob")
    print(p1)

    p2 = Player("robert")
    print(p2)
       
```
