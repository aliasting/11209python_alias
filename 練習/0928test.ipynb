```python
import random
class Player:
    def __init__(self,n:str):
        self.name = n
        self.__dice1 = random.randint(1,6)
        self.__dice2 = random.randint(1,6)
        self.__dice3 = random.randint(1,6)
        self.__dice4 = random.randint(1,6)

    #method l
    def __play(self) -> int:
        while True:
            dice = [self.__dice1,self.__dice2,self.__dice3,self.__dice4]
            if dice[0] == dice[1] == dice[2] == dice[3]:   
                score = 12 + self.__dice1
            elif dice[0] == dice[1] == dice[2] or dice[0] == dice[1] == dice[3]:   
                score = 0
            elif dice[0] == dice[2] == dice[3] or dice[1] == dice[2] == dice[3]:   
                score = 0
            
            elif dice[0] == dice[1] and dice[2] == dice[3]:    
                if dice[0] > dice[2]:
                    score = dice[0] + dice[1]
                else:
                    score = dice[2] + dice[3]
            elif dice[0] == dice[2] and dice[1] == dice[3]:    
                if dice[0] > dice[1]:
                    score = dice[0] + dice[2]
                else:
                    score = dice[1] + dice[3]

            elif dice[0] == dice[1]:    
                score = dice[2] + dice[3]
            elif dice[0] == dice[2]:    
                score = dice[1] + dice[3]
            elif dice[0] == dice[3]:    
                score = dice[1] + dice[2]
            elif dice[1] == dice[2]:    
                score = dice[0] + dice[3]
            elif dice[1] == dice[3]:    
                score = dice[0] + dice[2]
            elif dice[2] == dice[3]:    
                score = dice[0] + dice[1]
            elif not (dice[0] == dice[1] == dice[2] == dice[3]):   
                score = 0
            return score

    @property
    def value(self) -> int:
        return self.__play()
        
    def __repr__(self) -> str:
        descript = ""
        descript += f"{self.name}\n"
        descript += f"骰子1={self.__dice1}\n"
        descript += f"骰子2={self.__dice2}\n"
        descript += f"骰子3={self.__dice3}\n"
        descript += f"骰子4={self.__dice4}\n"
        descript += f"得分={self.value}分\n"
        return descript
    
    def __repr__(self) -> str:
        descript = f"姓名:{self.name}"
        return descript
    
def main():
    p1 = Player("jay")
    p2 = Player("John")

    while p1.value == 0:
        p1.roll_dice()
    while p2.value == 0:
        p2.roll_dice()

    print(p1)
    print(p2)

    print("遊戲结束")

if __name__ == "__main__":
   pass
    
```
